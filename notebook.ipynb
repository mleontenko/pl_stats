# PL Goalscoring Stats

### Import data

Import modules:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Import data:

```python
df = pd.read_csv('Premier League Player Stats.csv')
df.head()
```

Data info:

```python
df.info()
```

Data describe:

```python
df.describe()
```

### Analysis

Correlation matrix:

```python
corrMatrix = df.corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

Filter players with 10 or more goals:

```python
df = df[df["G"] >= 10]
df
```

Calculate minutes per goal (MPG):

```python
df["MPG"] = df["MIN"] / df["G"]
df.sort_values(by=['MPG'], inplace=True)
df = df.head(5)
df
```

```python
plt.figure(figsize=(15,10))

sns.scatterplot(x="G",
                    y="MIN",
                    data=df,
                    hue='PLAYER')
plt.show()
```
